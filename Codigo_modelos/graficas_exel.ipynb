```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de cada modelo desde los archivos de Excel
unet_data = pd.read_csv('/home/cornelius/Documentos/Modelos_tesis/output/training_logs_Unet_L5_L8_C2L1_P1.csv').dropna()
resunet_data = pd.read_csv('/home/cornelius/Documentos/Modelos_tesis/output/training_logs_DeepResUnet_L5_L8_C2L1_P1.csv').dropna()
deeplabv3_data = pd.read_csv('/home/cornelius/Documentos/Modelos_tesis/output/training_logs_DeepLabV3PLus_L5_L8_C2L1_P1.csv').dropna()

# Verificar la longitud de cada archivo despus de eliminar filas vacas
print("Longitud de datos Unet:", len(unet_data))
print("Longitud de datos ResUnet:", len(resunet_data))
print("Longitud de datos DeepLabV3:", len(deeplabv3_data))

# Seleccionar los datos de "Epoch" y "Avg MIoU Val"
epoch = unet_data['Epoch']  # Epoch constante
avg_miou_val_unet = unet_data['Avg MIoU Val']
avg_miou_val_resunet = resunet_data['Avg MIoU Val']
avg_miou_val_deeplabv3 = deeplabv3_data['Avg MIoU Val']

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(epoch, avg_miou_val_unet, label='U-Net', marker='o')
plt.plot(epoch, avg_miou_val_resunet, label='DeepResUNet', marker='s')
plt.plot(epoch, avg_miou_val_deeplabv3, label='DeepLabV3Plus', marker='^')

# Personalizacin de la grfica
plt.xlabel('Epoch')
plt.ylabel('MIoU')
#plt.title('Comportamiento de Avg MIoU Val en funcin de Epoch para diferentes modelos')
plt.legend(loc='lower right')  # Leyenda en la esquina inferior derecha
plt.grid(True)
plt.savefig('Miou_vs_epochs.png', bbox_inches='tight', dpi=300)
plt.show()
```
